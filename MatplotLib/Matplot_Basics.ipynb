```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([2,4,6,4]) # We didn't give X axis, So it has taken from 0
plt.title('Test Plot')
plt.xlabel('Indices')
plt.ylabel('Value')
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16]) 
plt.title('Squares')
plt.xlabel('Indices')
plt.ylabel('Square')
plt.grid()
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro') #Red Dot with 'o' symbol
plt.title('Squares')
plt.xlabel('Indices')
plt.ylabel('Square')
plt.grid()
plt.show()
```

```python
a=np.arange(4)
plt.plot(a,a*2,'r--',label='Double')
plt.plot(a,a**2,'bs',label='Squares')
plt.plot(a,a**3,'g--',label='Cubes')
plt.grid()
plt.legend()
plt.show()
```

# Controlling Different Lines

```python
x=np.array([1,2,3,4])
y=np.array([1,4,9,16])
plt.plot(x,y,linewidth=5.0) #Linewidth shows the thickness
plt.show()
```

```python
x1=np.array([1,2,3,4])
x2=np.array([1,2,4,8])
y1=np.array([1,4,9,16])
y2=np.array([1,8,27,64])

lines=plt.plot(x1,y1,x2,y2)
plt.setp(lines[0],color='r',linewidth=2.0)
plt.setp(lines[1],color='g',linewidth=3.0)
plt.show()
```

# Working with Multiple Figure and Axes

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

t1= np.arange(0,5,0.1)
t2= np.arange(0,5,0.02)

plt.subplot(211) #Creating 2 subplots. In 1st Subplot
plt.grid()
plt.plot(t1,f(t1),'b--')

plt.subplot(212) #In 2nd subplot
plt.grid()
plt.plot(t2,np.cos(2*np.pi*t2),'r*')

plt.show()
```
